```python
polarity = []
for i in data['comment_text']:
    try:
        polarity.append(TextBlob(i).sentiment.polarity)
    except:
        polarity.append(0)
```

```python
#checks whether punctution in tags and title have any relation with views, likes, dislikes comments
#like if punctutaion is increasing views is ncresing or noot
```

```python
import string
```

```python
string.punctuation
```

```python
def punc_counts(text):
    return len([char for char in text if char in string.punctuation])
```

```python
sample = full_df[0:1000]
```

```python
sample['count_punc'] = sample['title'].apply(punc_counts)
```

```python
sample['count_punc']
```

```python
plt.figure(figsize = (8,6))
sn.boxplot(x = 'count_punc',y = 'likes',data = sample)
plt.xticks(rotation  = 'vertical')
plt.show()
```
