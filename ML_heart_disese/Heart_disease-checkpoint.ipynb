```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
'exec(%matplotlib inline)' #
# It ensures that plots generated by matplotlib are displayed directly inside the notebook, not in a separate window.
import scipy.optimize as opt #
# Why we import it: If your code involves optimization problems (e.g., curve fitting, minimizing a function, etc.), you'll need this module.
import statsmodels.api as sm #statsmodels is a library for statistical modeling and hypothesis testing.
import matplotlib.mlab as mlab #Why we import it: You might use it for creating Gaussian (normal) distributions, signal processing, or other advanced visualizations.
import seaborn as sns
```

```python
data = pd.read_csv("ML_heart_disease.csv")
data.info()
```

```python
#coronary heart disease(CHD)
```

```python
data.drop(columns = ["education"],inplace = True,axis = 1)
data
```

```python
data.dropna(axis = 0,inplace = True)
data.isnull().sum()
data.describe()
```

```python
data["TenYearCHD"].value_counts()
X = np.asarray(data[['age', 'male', 'cigsPerDay', 
                           'totChol', 'sysBP', 'glucose']])
y = np.asarray(data['TenYearCHD'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=4)
# If X_train or X_test is a 1D array or scalar, reshape it to 2D
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# fitting logistic regression

```python
import warnings
# Ignore specific FutureWarnings
warnings.filterwarnings("ignore", category=FutureWarning)

from sklearn.linear_model import LogisticRegression
logistic =  LogisticRegression()
logistic = logistic.fit(X_train,y_train)
y_pred = logistic.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy of the model is =', 
      accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy of the model is =', 
      accuracy_score(y_test, y_pred))
```

# as we see that logistic regression is not good for this data set because in confusion matrix we geting high false negative  so we will se from other model also



```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
models = {
    'gnb' :GaussianNB(),
    'Rf' :RandomForestClassifier(random_state = 42),
    'Dt' :DecisionTreeClassifier(random_state = 42),
    'Ab' : AdaBoostClassifier(n_estimators=100, random_state=0),
    'clf' : GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,
    max_depth=1, random_state=0),
    
}
for model_name in models:
    model = models[model_name]
    scores = cross_val_score(model,X_train,y_train,cv = 10,scoring = 'accuracy')
    print("=="*30)
    print(model)
    print(f"score:{scores}")
    print(f"Mean Score: {scores.mean()}")
```

```python
final_ad =  AdaBoostClassifier(n_estimators=50, random_state=42)
# Number of small models AdaBoost will use to create a strong model.
#A higher value often means better accuracy, but it also means more time to train.
#If n_estimators=3, it’s like having 3 small decision trees:
final_ad = final_ad.fit(X_train,y_train)
y_preds = final_ad.predict(X_test)
accuracy_score(y_test, y_preds)
```

```python
cm = confusion_matrix(y_test,y_preds)
cm
X_test
```
